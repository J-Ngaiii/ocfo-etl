```python
import pandas as pd
import re
```

```python
raw = pd.read_csv('experimental_data/FinancialTransactionList.csv', header=1)
```

```python
raw.head(5)
```

```python
raw['Category'].value_counts()
```

```python
def not_admin(entry: str):
    ENTRY = entry.lower()
    if 'admin' in ENTRY:
        return False
    elif 'internal transfer' in ENTRY or 'transfer between accounts' in ENTRY:
        return False
    else:
        return True
```

```python
tbl = raw[raw['Category'].astype(str).apply(not_admin)]
sum(
    tbl['Amount'].apply(
        lambda x: float(re.match(r'^\$(.*)', x).group(1))
    )
)
```
